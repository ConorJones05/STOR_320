#### 6 Compare with uniform distribution.

- Extract all values from `df_cumsum` that are less than 10 and flatten them into a one-dimensional array named `cumsum_values_lt_10`. (5 points)

- Plot the density distribution of `cumsum_values_lt_10` using `sns.histplot`. (5 points)

```python
# Extract all values from `df_cumsum` that are less than 10 and flatten them into a one-dimensional array named `cumsum_values_lt_10`. (5 points)
cumsum_values_lt_10 = df_cumsum[df_cumsum < 10].values.flatten()
cumsum_values_lt_10
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Plot the density distribution of `cumsum_values_lt_10` using `sns.histplot`
sns.histplot(cumsum_values_lt_10, color='blue', kde=True)
plt.show()
```

```python
# Generate a numpy array named `uniform_values` consisting of 100,000 variables that follow the  uniform distribution between 
# [0, 10]. On the same figure, plot the density distribution of `uniform_values`. (10 points)
uniform_values = np.random.uniform(0, 10, 100000)
sns.histplot(cumsum_values_lt_10, color='blue', label='Cumsum Values',kde=True)
sns.histplot(uniform_values, color='pink', label='Uniform Values', kde=True)
plt.legend()
plt.show()
```

The distribution of cumsum_values_lt_10 and uniform_values are similar in shape, but they exhibit diffrent characteristics. The cumulative sum values generated from the exponential distribution tend to cluster more around smaller values where shorter intervals between events are more common. In contrast, the uniform distribution shows an even spread of values between 0 and 10. This highlights the difference between an exponentially decaying process and a uniformly distributed process.

#### 7. Create `max_indices` (10 points)

- For each column in `df_cumsum`, find the maximum row index where the cumulative sum is less than or equal to 10.

- Create a Series named `max_indices` containing these indices for all 1,000 columns.

```python
max_indices_dict = {}

for col in df_cumsum.columns:
    indices = df_cumsum[df_cumsum[col] <= 10].index
    max_indices_dict[col] = indices.max()

max_indices = pd.Series(max_indices_dict)
max_indices
```
